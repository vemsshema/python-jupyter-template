### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.arange(1, 6)
y = np.array([10, 20, 15, 25, 30])
bar_heights = [3, 7, 8, 5, 6]
scatter_y = np.array([5, 15, 10, 20, 25])
labels = ['A', 'B', 'C', 'D']
sizes = [40, 30, 20, 10]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart (Top-left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Chart')

# Bar chart (Top-right)
axs[0, 1].bar(x, bar_heights, color='g')
axs[0, 1].set_title('Bar Chart')

# Scatter plot (Bottom-left)
axs[1, 0].scatter(x, scatter_y, color='r', marker='x')
axs[1, 0].set_title('Scatter Plot')

# Pie chart (Bottom-right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
