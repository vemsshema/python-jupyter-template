### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

def plot_bar_chart(categories, values):
    # Ensure the lengths match
    if len(categories) != len(values):
        raise ValueError("The number of categories and values must be the same.")
    
    # Create bar plot
    plt.figure(figsize=(6, 4))
    plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
    
    # Labels and title
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.title("Bar Plot of Given Data")
    
    # Show plot
    plt.show()

# Example data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Call function
plot_bar_chart(categories, values)
```
